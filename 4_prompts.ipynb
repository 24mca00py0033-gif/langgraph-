```python
%pip install python-dotenv langchain-groq langgraph
from dotenv import load_dotenv
import os

load_dotenv()

if os.getenv("GROQ_API_KEY"):
    print("Groq API key is set.")
else:
    raise ValueError("Groq API key is not set.")
from langchain_groq import ChatGroq

model = ChatGroq(
    model="llama-3.3-70b-versatile"
)
model
```

```python
from langchain_core.prompts import ChatPromptTemplate
my_prompt=ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that helps to create social media post in genz tone and funny "),
    ("human", "Create a social media post about {topic}.")
])
```

```python
chain=my_prompt | model
chain.invoke({"topic":"the new iPhone 15"}).content
```
